```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
import os
# Get the parent directory of the current file (notebook)
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
# Get the parent directory of the current file (notebook)
disorder_df = pd.read_csv(f"{parent_dir}/Data/processed_disorder_scores.csv")
```

```python
disorder_df.head()
```

```python
disorder_df = disorder_df.astype('object')
```

```python
for column in disorder_df.columns[:-1]:
    plt.figure(figsize=(10, 6))  # Set the figure size
    disorder_df[column].value_counts().plot(kind='bar')
    plt.title(f'Bar Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```
