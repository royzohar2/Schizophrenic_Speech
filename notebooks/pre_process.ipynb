# word pred llm


```python
file_path = f"{parent_dir}/data/word_pred_llm.json"
with open(file_path, 'r', encoding = 'utf-8') as f:
    data_js = json.load(f)
```

```python
df_rows = []
for item in data_js:
    file_name = item['file_name']
    label = item['label']
    
    for question, vector in item['questions'].items():
        mean_value = np.mean(vector[1:])  # Calculate mean of the vector
        df_rows.append({
            'file_name': file_name,
            'label': label,
            'question': question,
            'mean_vector': mean_value
        })

word_pred_llm_df = pd.DataFrame(df_rows)
word_pred_llm_df.head()
```

```python
word_pred_llm_df.shape
```

```python
word_pred_llm_df.rename(columns={"file_name": "person"}, inplace=True)
word_pred_llm_df.head()
```

# tf_idf

```python
tf_idf_df = pd.read_csv(f"{parent_dir}/data/morphological_yap_features.csv", index_col=False)
tf_idf_df = tf_idf_df[["question", "label", "person", "tf_idf_seg_txt", "tf_idf_tok_txt", "count_vec"]]
tf_idf_df.head()
```

```python
# Function to calculate the mean of the numeric part in each string list
def calculate_mean_from_list_string(list_string):
    try:
        parsed_list = ast.literal_eval(list_string)
        numeric_values = [float(item[1]) for item in parsed_list]
        return sum(numeric_values) / len(numeric_values) if numeric_values else None
    except (SyntaxError, ValueError):
        return None

# Apply the function to each column of interest
tf_idf_df['mean_tf_idf_seg_txt'] = tf_idf_df['tf_idf_seg_txt'].apply(calculate_mean_from_list_string)
```

```python
tf_idf_df_filtered = tf_idf_df[['person' ,'label' ,"question" ,"mean_tf_idf_seg_txt"]]
tf_idf_df_filtered.head()
```

```python
raw_data = pd.read_csv(f"{parent_dir}/data/clean_data.csv")
raw_data.head(1)
```

```python
raw_data.shape 
```

```python
# Define the columns to reshape
image_columns = ['תמונה 1', 'תמונה 2', 'תמונה 3', 'תמונה 4', 'תמונה 5', 'תמונה 6', 
                 'תמונה 7', 'תמונה 8', 'תמונה 9', 'תמונה 10', 'תמונה 11', 'תמונה 12', 
                 'תמונה 13', 'תמונה 14' ,'בר מצווה', 'מה אוהב לעשות', 'מה מעצבן', 'משאלות לעתיד']


# Reshape the DataFrame
reshaped_df = raw_data.melt(
    id_vars=['file_name', 'label'],  # Columns that remain the same for each row
    value_vars=image_columns,     # Columns to melt (the image columns)
    var_name='question',          # Column name for the image identifier (e.g., תמונה 1)
    value_name='answer'           # Column name for the image descriptions
)

reshaped_df.rename(columns={"file_name": "person"}, inplace=True)

# Display the reshaped DataFrame
print(reshaped_df.shape)
reshaped_df.head(1)
```

```python
# Find all rows where 'answer' is null or empty
empty_or_null_answers = reshaped_df[reshaped_df['answer'].isnull() | reshaped_df['answer'].str.strip().eq('') | reshaped_df['answer'].str.strip().eq('[]')]

# Display the rows with empty or null answers
empty_or_null_answers.head(50)
```

```python
# Merge the first two DataFrames with an outer join on 'person', 'label', and 'question'
merged_df = pd.merge(scores_per_sentence_df_filtered, word_pred_llm_df, on=['person', 'label', 'question'], how='outer')

# Merge the resulting DataFrame with the third DataFrame with an outer join on 'person', 'label', and 'question'
final_merged_df = pd.merge(merged_df, tf_idf_df_filtered, on=['person', 'label', 'question'], how='outer')

# Display the result
final_merged_df.head()
```

```python
final_merged_df.shape
```

```python
# all the person + question that in empty_or_null_answers remove from final_merged_df
# Extract 'person' and 'question' pairs from empty_or_null_answers
pairs_to_remove = empty_or_null_answers[['person', 'question']]

# Merge the pairs to remove with final_merged_df to get rows to exclude
filtered_final_merged_df = final_merged_df.merge(pairs_to_remove, on=['person', 'question'], how='left', indicator=True)

# Filter out rows where the merge indicator is 'both', meaning the pair is in empty_or_null_answers
filtered_final_merged_df = filtered_final_merged_df[filtered_final_merged_df['_merge'] == 'left_only'].drop(columns='_merge')

# Display the filtered DataFrame
filtered_final_merged_df.head(1)
```

```python
# Total NaN values in the entire DataFrame
total_nulls = filtered_final_merged_df.isnull().sum().sum()

# Display the total count of NaN values
print(f"Total NaN values in the DataFrame: {total_nulls}")

```

```python
# Count of NaN values in each column
nulls_per_column = filtered_final_merged_df.isnull().sum()

# Display the count of NaN values per column
print(nulls_per_column)
```

```python
# Count of NaN values in each row
nulls_per_row = filtered_final_merged_df.isnull().sum(axis=1).sum()

# Display the count of NaN values per row
print(nulls_per_row)
```

```python
# Calculate the mean of 'mean_tf_idf_seg_txt', excluding NaNs
mean_tf_idf = filtered_final_merged_df['mean_tf_idf_seg_txt'].mean()

# Replace NaN values in 'mean_tf_idf_seg_txt' with the calculated mean
filtered_final_merged_df['mean_tf_idf_seg_txt'].fillna(mean_tf_idf, inplace=True)

# Count of NaN values in each column
nulls_per_column = filtered_final_merged_df.isnull().sum()

# Display the count of NaN values per column
print(nulls_per_column)
```
